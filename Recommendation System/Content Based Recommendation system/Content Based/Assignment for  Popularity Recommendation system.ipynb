```python
import pandas as pd

def top5_movies_by_genre(items_file, ratings_file, genres=None):
    # Define the column names
    i_cols = ['movie id', 'movie title', 'release date', 'video release date', 'IMDb URL', 'unknown', 'Action', 'Adventure',
              'Animation', "Children's", 'Comedy', 'Crime', 'Documentary', 'Drama', 'Fantasy', 'Film-Noir', 'Horror',
              'Musical', 'Mystery', 'Romance', 'Sci-Fi', 'Thriller', 'War', 'Western']

    # Load the items dataset
    items = pd.read_csv(items_file, sep='|', names=i_cols, encoding='latin-1')

    # If genres are specified, find top 5 movies for each genre
    if genres:
        # Load the ratings dataset
        ratings_cols = ['user id', 'movie id', 'rating', 'timestamp']
        ratings = pd.read_csv(ratings_file, sep='\t', names=ratings_cols)

        # Merge items and ratings datasets
        merged_data = pd.merge(items[['movie id', 'movie title']], ratings[['movie id', 'rating']], on='movie id')

        # Compute the average rating for each movie within each genre
        genre_avg_ratings = merged_data.groupby('movie title')['rating'].mean().reset_index()

        # Initialize an empty dictionary to store the top 5 highest-rated movies for each genre
        top5_movies_by_genre = {}

        # Loop through each genre
        for genre in genres:
            # Filter movies by the current genre
            genre_movies = items[items[genre] == 1]

            # Merge genre_movies with genre_avg_ratings
            genre_merged = pd.merge(genre_movies[['movie id', 'movie title']], genre_avg_ratings, on='movie title')

            # Sort the genre_merged dataframe by rating in descending order
            sorted_genre = genre_merged.sort_values(by='rating', ascending=False)

            # Select the top 5 highest-rated movies for the current genre
            top5_movies_by_genre[genre] = sorted_genre.head(5)

        return top5_movies_by_genre
    
    # If no genres specified, find overall top 5 movies
    else:
        return items[['movie title']].head(5)

# Define the file paths for items and ratings datasets
items_file_path = 'ml-100k/u.item'
ratings_file_path = 'ml-100k/u.data'

# Define the genres for which you want to find the top 5 highest-rated movies
genres_to_check =  ['movie id', 'movie title', 'release date', 'video release date', 'IMDb URL', 'unknown', 'Action', 'Adventure',
              'Animation', "Children's", 'Comedy', 'Crime', 'Documentary', 'Drama', 'Fantasy', 'Film-Noir', 'Horror',
              'Musical', 'Mystery', 'Romance', 'Sci-Fi', 'Thriller', 'War', 'Western']
  # Example genres

# Call the function and get the top 5 highest-rated movies for each genre
top5_movies = top5_movies_by_genre(items_file_path, ratings_file_path, genres_to_check)

# Print the top 5 highest-rated movies for each genre
for genre, top5_movies_list in top5_movies.items():
    print(f"\nTop 5 Highest-Rated {genre} Movies:")
    for index, row in top5_movies_list.iterrows():
        print(f"{row['movie title']}: {row['rating']}")
```
